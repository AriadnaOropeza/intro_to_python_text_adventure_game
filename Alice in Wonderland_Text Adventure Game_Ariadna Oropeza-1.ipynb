```python
## Text Adventure Game
## Data Science: Python
## Hult International Business School

## Alice in Wonderland
## source: https://www.imdb.com/title/tt0043274/plotsummary

"""
@author: Ariadna Donaji Oropeza Mu√±iz

"""
"""
docstring

Welcome to the amazing world of Alice in Wonderland! Or is it? Through this
game you will be able to discover that! Based on the 1951 Disney movie, 
you will get to discover and wonder through the different environments that
make up this world, and mingle with different fascinating creatures and 
people. 
Beware though, you have to make it back to the real world before the Queen 
of Hearts yells OFF WITH HER HEAD!
    
Have fun!

"""

import random
import sys

#First scene of the movie.
#Defining the setting for the user.

def living_room():
    print(f"""
    -Start-
    {'_' * 50}
    
    You are in the living room with your sister, who is reading about William I of England.
    You are bored and daydreaming of a nonsense world, when you see a rabbit hopping by
    wearing a vest and mumbling something to his pocket watch. This cannot be a coincidence.
    You leave your sister and decide to follow him and discover the world he comes from.\n""")
    input(prompt = " <Press enter to continue >\n")
    rabbit_hole()
    
#The first stage of the game.
#Following the rabbit and entering Wonderland. 

def rabbit_hole():
    
    print("""
    You go through a rabbit hole, which leads to a door with a miniature door. There is a potion
    which says 'drink me' and a cookie jar labelled 'eat me'. Which one do you pick?
    
    """)
    
# The user is expected to choose one of the two options provided for futher action within the game. 
# Using conditional statements, the previous action of the user will either advance him-her in the game
# or give them a set of random questions to answer to move forward to the next stage. 

    choice = input(prompt = " 1. drink from the water bottle or 2. take a bite from a cookie\n")
    choice = choice.lower()
    
    if "1" in choice or "drink" in choice or "water" in choice:
        
        print("""\n You shrunk to a third of your size! But you didn't grab the key before drinking.
        Now, you need help from the doorknob to get through the door without it. Just take a quiz and
        he will let you through""")
        
        
        question_list = ['Boston campus dean?',
                        'Largest mammal in the world?',
                        'Largest continent?',
                        'Most spoken language in the world?',
                        'Number of countries in the world?',
                        'First woman to win a Nobel Prize?']
        questions = 3
        
# Using a while loop, the user will have to correctly answer three questions to be able to move to the next stage.
# While the questions answered are > 0, the user will continue to receive a random question to answer. 
# Once three questions are answered, the loop will break and the user will either move to the next stage or
# they will be given the opportunity for a new chance in the game. 
        
        while questions > 0:
            question = random.choice(question_list)
            input(prompt = "< Press enter for your question>\n")
            print(question)
            print("")
            
# For the following, the user will get a random question to answer from the list.

            if question == question_list[0]:
                print(" a) Dean Chavez")
                print(" b) Dean Heaton")
                print(" c) Dean Peazant")
                solution_1 = input(prompt = '>\n')
                solution_1 = solution_1.lower()
                
                if "a" in solution_1 or "A" in solution_1 or "Chavez" in solution_1 or "chavez" in solution_1:
                    print("That's correct!")
                    questions -=1
                    
                else:
                    print("That is incorrect!")
                    another_chance()
                    
                    
            elif question == question_list[1]:
                print("a) elephant")
                print("b) antartic blue whale")
                print("c) hippo")
                solution_2 = input(prompt = '>\n')
                solution_2 = solution_2.lower()
                
                if "b" in solution_2 or "whale" in solution_2 or "antartic" in solution_2:
                    print("That is correct!")
                    questions -=1
                    
                else:
                    print("That is incorrect!")
                    another_chance()
                        
            elif question == question_list[2]:
                print("a) North America")
                print("b) Africa")
                print("c) Asia")
                solution_3 = input(prompt = '>\n')
                solution_3 = solution_3.lower()
                
                if "c" in solution_3 or "C" in solution_3 or "Asia" in solution_3 or "asia" in solution_3:
                    print("Excellent! You know your geography!")
                    questions -=1
                    
                else:
                    print("Sorry, that's a no")
                    another_chance()
                    
            elif question == question_list[3]: 
                print("a) Chinese/Mandarin")
                print("b) English")
                print("c) Hindi")
                solution_4 = input(prompt = '>\n')
                solution_4 = solution_4.lower()
                    
                if "b" in solution_4 or "B" in solution_4 or "english" in solution_4 or "English" in solution_4:
                    print("That is correct!")
                    questions -=1
                    
                else:
                    print("That is incorrect!")
                    another_chance()
                    
            elif question == question_list[4]:
                print(" a) There are 192 countries")
                print(" b) There are 195 countries")
                print(" c) There are 190 countries")
                solution_5 = input(prompt = '>\n')
                solution_5 = solution_5.lower()
                    
                if "195" in solution_5 or "b" in solution_5 or "B" in solution_5:
                    print("That is correct!")
                    questions -=1
                    
                elif "192" in solution_5 or "190" in solution_5 or "a" in solution_5 or "c" in solution_5:
                    print("Nope. Seems like you should go back to school!")
                    another_chance()
            
            elif question == question_list[5]:
                print(" a) Marie Curie")
                print(" b) Rosalind Franklin")
                print(" c) Lise Meitner")
                solution_6 = input(prompt = '>\n')
                solution_6 = solution_6.lower()
                
                if "a" in solution_6 or "A" in solution_6 or "Marie" in solution_6:
                    print("That is correct!")
                    questions -=1
                    
                else:
                    print("That is incorrect, sorry!")
                    another_chance()
                    
                    break
                    

        print(input(prompt = f"""And through the door you go! Go find out what there is behind it!

        < press enter to continue walking> \n"""))
        hookah_caterpillar()
        
# If the user decides to eat a cookie, they will be given a guessing game.
# The guessing game is done with the random.randint function, so they have to guess a number 1-15.

    elif "2" in choice or "take" in choice or "bite" in choice or "cookie" in choice:
        
        print(""""Wow! You are huge now! Don't worry though, let's play a guessing game
        and I'll tell you how to get the perfect size to go through the door!\n""")
        
        print('Are you able to guess the number I am thinking about right now?\n')
        guess = input(prompt = 'I will help you out. Just pick a number from 1 to 15\n')
        guess= int(guess)
        number=random.randint(a = 1,
                              b = 15)
        print(f"""
        Tell me... did you guess it (True/False)?
        {guess == number}
        
        The number I was thinking of was {number}!
        """)
        
        hookah_caterpillar()

    else:
        print("That's not a correct input, sorry! Try again\n")
        input(prompt = "< Press enter to continue >")
        rabbit_hole()
        
##Second stage of the game.

# You walk through the door and enter a new environment within Wonderland.
# In this stage, the user has to guess a color, and it has three chances to get it right. 

def hookah_caterpillar():
    print(f"""
    
    {'_' * 50}
    
    After passing a garden full with rude roses, you encounter a caterpillar, who is smoking from a hookah.
    He is about to turn into a butterfly, but you need his help to keep going. The only way he will help 
    you out, is if you guess the color of his future wings!\n""")
    
    input(prompt = "< Press enter to continue >\n")
    
    wings_color = 'blue'
    guess = ""
    guess = guess.lower()
    chances = 0
    
# A while loop was used here, when the user guesses the color they move to the next stage
# If they run out of chances, the game ends with a fail() function.

    while guess != wings_color:
        if chances <= 3:
            guess = input(prompt = " What color will my wings be?\n")
            guess = guess.lower()
            chances += 1
            
            if guess == wings_color:
                mad_hatter()
        
        else:
            print("Your tries are up! As well is the lovely blue winged butterfly that has just left")
            fail()

            
## Third stage of the game

# In this stage, the user has to answer a question through conditional statements. 
# If answered correctly, the user moves to the fourth and final stage of the game. 
    
def mad_hatter():
    print(f"""
    
    {'_' * 50}
    
    Before flying away, the butterfly gave you some directions on which way the rabbit went.
    On your way, you get distracted by an odd looking house, where you find the March Hare and Mad Hatter
    celebrating their unbirthday!\n""")

    
    input(prompt = "< Press enter to continue and join them >\n")
    
    print(f"""
    
    {'_' * 50}
    
    You join their celebration! However, they are a little odd, and you are still 
    looking for the rabbit. The only way you will be able to leave is if you
    answer the Mad Hatter's riddle correctly:
    What question can you never answer yes to?
    
    1. Are you hungry?
    2. Are you asleep yet?
    3. Are you tired now? \t""")
    
    riddle_answer = input(prompt = "Your answer is:\n")
    riddle_answer = riddle_answer.lower()
    
    if "2" in riddle_answer or "Are you asleep yet?" in riddle_answer or "asleep" in riddle_answer:
        
        print("WOOF! You got out of there! Keep on walking, you'll reach the rabbit at some point\n")
        
        queen_of_hearts()
        
# If the answer is wrong, the user doesn't move forward but has the option to start the game again.        
        
    else:
        print(f"""
        That is incorrect! Seems like you will be staying with them now!\n""")
        another_chance()
        
## Fourth and last stage of the game

# In this stage a while loop is used as well as a random.choice function
# for the while loop, the user has to press the enter key three times for it to end. 

def queen_of_hearts():
    print(f"""
    
    {'_' * 50}
    
          The rabbit entered the Queen of Hearts garden, where playing cards are
          painting white roses, red. 
          You get invited by the queen to a game of croquet, where flamingos are mallets
          and hedgehogs the balls. Everyone is terrified of her, but you accept to play\n""")
    
    input(prompt="<Press enter to continue and start playing>\n")
    presses = 4
    
    while presses > 0:
        input(prompt = "< The queen has played! Your turn. Press enter to score >\n")
        presses -= 1

# After the while loop, a random.choice is used for the user to play a 'guessing game' with the queen. 
# It does not matter if the user guesses correctly or not, it is only a challenge set by the queen. 
              
    print("Before you go, the Queen would like to ask you to play a guessing game with her. Are you ready?")
    
    guessing_list = ['rabbit', 'mad hatter', 'chesire cat', 'smoking hookah caterpillar']
    guessing_game = input(prompt= """I bet you cannot tell me the word I am thinking about just now
                     You have met this character in your journey through Wonderland\n""")
        
    correct_guess = random.choice(guessing_list)
        
    print(f"""
        Tell me now... what is the word I thought of? Did you guess it right (True\False)? 
        {guessing_game == correct_guess}
        
        The word I was thinking of was {correct_guess}!
        
        """)

    
    print(f"""
        Yay! You played and won the game fair and square. The Queen likes a good rival, your head is safe.
        As you score the last shot, you wake up and realize it has all been an amazing dream.\n""")
    another_chance()
    

# The another_chance function gives the user the opportunity to play again when they have come to the end of a round. 
# The end of a round can be because of mistakes or because the game ended in the fourth stage. 
# If the user types 'yes', they will go to the rabbit_hole. 
# If the user types 'no' then, they will get a message thanking them for playing the game. 
    
def another_chance():
    another_chance = input(prompt = "Do you want to have another round? 'yes' or 'no'?\n")
    another_chance = another_chance.lower()
    if "yes" in another_chance or "Yes" in another_chance:
        rabbit_hole()
     
            
    else:
        input(prompt = "< Press enter to wake up and exit Wonderland >\n")
        print(f"""Thank you for your visit to this wonderful world! Bye!\n""")
        
# The restart function allows the user to enter Wonderland again by pressing enter.       
    
def restart():
    restart = input(prompt = "Into Wonderland we go again! PRESS ENTER ")
    rabbit_hole()

# The fail function is used whenever the user loses the game because of mistakes.
# They will receive an "Off with your head" message and a Game Over prompt will appear.
# The user will then be given the opportunity to play again, if that is what they want. 
    
def fail():
    print('Off with your head! That is game over for you.')
    input( prompt = '< GAME OVER >')
    exit(0)
    restart()
    rabbit_hole()

                                 
living_room() #this is the start function of the game       
```
